```python
import numpy.linalg as nl
import numpy as np

import scipy.io as sio

mat_contents = sio.loadmat('system_identification_eval.mat')

x0 = mat_contents['x0']
xd0 = mat_contents['xd0']+29.0576
xdp = mat_contents['xdp']+29.0576
xp = mat_contents['xp']
xp = np.transpose(xp)
xdp = np.transpose(xdp)
```

```python
time_interval = 1/15
def acceleration (x, xd, xp, xdp):
    return -0.01150047*x + -0.31465897 * xd + 0.01154398* xp + 0.28080008* xdp

def disp(u , t, a):
    return u*t +0.5*a *t *t

def velocity(u, t, a):
    return u + a*t
distances = list()
x = x0
distances.append(x)
xd = xd0
xdd = acceleration(x, xd, xp[0], xdp[0])

for t in range (1, 150):
    x = x + disp(xd, time_interval, xdd)
    xd = velocity(xd, time_interval, xdd)
    xdd = acceleration(x, xd, xp[t], xdp[t])
    distances.append(x)
distances
```

```python
np.savetxt('submission.txt', distances, newline='\r\n')
```
